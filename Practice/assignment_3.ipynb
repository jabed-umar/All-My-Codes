```python
# Cholesky decomposition method to solve linear equation
def Cholesky(A,b,x):
    # finding the number of rows
    n = len(A)
    # Finding the diagonal entites of L
    for i in range(n):
        sum = 0
        for k in range(i):
                sum += (A[i][k] * A[i][k])
                A[i][i] = (A[i][i] - sum)**(1/2)
# Finding the non-digonal entities of L
    for i in range(n):
        for j in range(i+1,n):
            sum = 0
            for k in range(i):
                    sum += (A[j][k] * A[i][k])
                    A[j][i] = (A[j][i] - sum)/A[i][i]
# forward substitution
    for i in range(n):
        sum = 0
        for j in range(i):
            sum += (A[i][j] * x[j])
            x[i] = b[i] - sum
    # backward substitution
    for i in range(n-1,-1,-1):
        sum = 0
        for j in range(i+1,n):
            sum += (A[i][j] * x[j])
            x[i] = (x[i] - sum)/A[i][i]
    return x
# main function
b = [2,1,2,1,2,1]
A = [
    [4, -1, 0, -1, 0, 0],
    [-1, 4, -1, 0, -1, 0],
    [0, -1, 4, 0, 0, -1],
    [-1, 0, 0, 4, -1, 0],
    [0, -1, 0, -1, 4, -1],
    [0, 0, -1, 0, -1, 4],
    ]
x = [0,0,0,0,0,0]
# calling the function
Cholesky(A,b,x)
print("The solution of the linear equation is : ",x)
```

```python
# Check the given matrix is symmetric or not 
def Symmetric(mat):
    """_summary_ : Check if a matrix is symmetric

    Args:
        mat (2-d array): sqaure matrix of order n (n>=2)

    Returns:
        bool : True if the matrix is symmetric, else False
    """
    N = len(mat)
    for i in range(N):
        for j in range(N):
            if (mat[i][j] != mat[j][i]):
                return False
    return True
```
